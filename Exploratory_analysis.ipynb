```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("vehicle_emissions.csv")

df.head()
```

```python
df.hist(figsize=(12,10))
plt.show()
```

```python
sns.histplot(df['CO2_Emissions'], kde=True)
plt.title("Distribution of CO2 Emissions")
```

```python
for col in df.select_dtypes(include='object').columns:
    print(f"\n{col}")
    print(df[col].value_counts())
```

```python
sns.scatterplot(x='Engine_Size', y='CO2_Emissions', data=df)
plt.title("Engine Size vs CO2 Emissions")
```

```python
sns.scatterplot(x='Fuel_Consumption_comb(L/100km)', y='CO2_Emissions', data=df)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

1. Engine size has a strong positive relationship with CO2 emissions.
2. Fuel consumption is the most influential factor affecting emissions.
3. Vehicles with more cylinders tend to emit higher CO2.
4. SUVs and trucks produce significantly higher emissions compared to compact cars.
5. There are noticeable outliers with extremely high emissions, indicating high-performance or heavy vehicles.

Conclusion : The EDA reveals that vehicle specifications such as engine size, fuel consumption, and cylinders are key drivers of CO2 emissions. These insights guided feature selection and preprocessing steps used in the machine learning model.
